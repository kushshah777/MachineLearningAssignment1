```
import numpy as np
import keras
from keras.datasets import mnist
from keras import backend as K
rows=28
columns=28
```

```
def activationsigmoid(x): #sigmoid is used as activation function here as we are interested to know if the input image is YES/NO belonging to a particular class (0-9)
  return 1.0/(1.0+ np.exp(-x))
def sigmoidprime(x): #sigmoid prime is used for calculating the derivatives of sigmoid function because at the output layer we have sigmoid as activation function so
  return activationsigmoid(x)*(1-activationsigmoid(-x))
def minibatches(X,Y,size): # used to calculate minibatches
  for i in range(0,X.shape[0]-size+1,size):
    a=slice(i,i+size)
    yield X[a],Y[a]
def softmaxfunction(x):
  x-=np.max(x)
  return np.exp(x) / np.sum(np.exp(x), axis=0)
def crossentropyloss(y,yHat):
  return -np.mean(y * np.log(yHat))
```

```
def neuralnetwork(X_train,Y_train,X_test,Y_test,epochs,learningrate):
  inputimage=X_train.shape[0]
  inputlabel=Y_train.shape[0]
  
  initialweight=np.random.randn(inputlabel,inputimage)
  initialbias=np.zeros(shape=(inputlabel,1))
  batchsize=32
  for i in range(epochs):
    for b in minibatches(X_train.T,Y_train.T,batchsize):
      x1,y1=b
      x1=x1.T
      y1=y1.T
      sample=x1.shape[0]
      trans1=np.dot(initialweight,x1) + initialbias
      output1=softmaxfunction(trans1)
      loss= crossentropyloss(y1,output1)
      dtrans1=output1-y1
      dinitialweight= np.dot(dtrans1,x1.T) * (1/sample)
      dinitialbias= np.sum(dtrans1, axis=1,keepdims=True) * (1/sample)
      initialweight= dinitialweight * initialweight-learningrate
      initialbias=dinitialbias * initialbias-learningrate
      
      print("epoch no ",i," has cost",loss)
  trans1=np.dot(initialweight,X_train) + initialbias
  output2=softmaxfunction(trans1)
  
  transtest=np.dot(initialweight,X_test)+ initialbias
  outputtest=softmaxfunction(transtest)
  
  outputtest = outputtest.T
  output2 = output2.T
  Y_train=Y_train.T
  Y_test=Y_test.T
  
  
  Y_train_predict=np.zeros_like(Y_train)
  Y_train_predict[np.arange(len(output2)),output2.argmax(1)] = 1
  
  Y_test_predict=np.zeros_like(Y_test)
  Y_test_predict[np.arange(len(outputtest)),outputtest.argmax(1)] =1
  
  trainaccuracy=Y_train_predict-Y_train
  testaccuracy=Y_test_predict-Y_test
  print(""+str(trainaccuracy))
  print(""+str(testaccuracy))
  array ={"loss": loss,"predictedtestlabel":outputtest, "predictedtrainlabel":output2}
  
```

```
(x_train,y_train), (x_test,y_test ) = mnist.load_data() # loading data from mnist dataset
rows=28
columns=28
if K.image_data_format() == 'channels_first':          #checking if the data format is channel first, if so we add 1 before rows and columns.
  x_train=x_train.reshape(x_train.shape[0],1,rows, columns)
  x_test=x_test.reshape(x_test.shape[0],1,rows,columns)
  inputshape=(1,rows,columns)
else:
  x_train=x_train.reshape(x_train.shape[0],rows,columns,1)
  x_test=x_test.reshape(x_test.shape[0],rows,columns,1)
  inputshape=(rows,columns,1)
```

```
x_train = x_train / 255.0     #normalizing the data
x_test = x_test / 255.0
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
x_train = x_train.reshape(x_train.shape[0], -1).T  #scaling the data to fit our neuralnetwork
x_test = x_test.reshape(x_test.shape[0], -1).T
y_test = y_test.T
y_train = y_train.T
```

```
model=neuralnetwork(x_train,y_train,x_test,y_test,10,0.1)
```
